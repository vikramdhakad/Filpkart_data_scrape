```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.flipkart.com/search?q=iphones&as=on&as-show=on&otracker=AS_Query_HistoryAutoSuggest_7_0_na_na_na&otracker1=AS_Query_HistoryAutoSuggest_7_0_na_na_na&as-pos=7&as-type=HISTORY&suggestionId=iphones&requestId=676a98e0-e01d-46b4-8b92-ed518e8a8c59"
req = requests.get(url)
```

```python
Name = []
Price = []
Description = []
```

```python
soup = BeautifulSoup(req.content, "html.parser")
```

```python
nam = soup.find_all("div", class_="_4rR01T")
```

```python
for i in nam:
    Name.append(i.text)
    
Name
```

```python
pric = soup.find_all("div", class_='_30jeq3 _1_WHN1')
```

```python
for i in pric:
    Price.append(i.text)
    
Price
```

```python
des = soup.find_all("div", class_='fMghEO')
```

```python
for i in des:
    Description.append(i.text)
    
Description
```

```python
data_file = pd.DataFrame()
data_file['Product_Name'] = Name
data_file['Product_Price'] = Price
data_file['Info'] = Description
data_file
```

```python
data_file.to_csv('Product_data.csv', index=False)
```
